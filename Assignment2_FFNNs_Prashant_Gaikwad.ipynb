## Preparing Data for Modeling

### Subsetting the data for predictor and response variables 

```python
# Setting up the predictor variable in Y
Y = df['MWh']

# Setting up the response/ dependent variables in X
X_1 = df.loc[:,'tmpf':'sknt']
X_2 = df.loc[:,'HE']
X = [X_1,X_2]
X = pd.concat(X, axis = 1)
```

```python
# Looking at the X 
X.head()
```

```python
# Looking at the Shape of X and Y
print(X.shape, Y.shape)
```

### Splitting the Data in Test and Training set

```python
# convert to numpy array
X = np.array(X)
y = np.array(Y)
```

```python
# splitting Using the module method
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, shuffle = True, random_state = 2994408)
```

```python
# Looking at the shape of each split
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

### Scaling the splitted data

```python
# use minMax scaler
min_max_scaler = MinMaxScaler()
X_train = min_max_scaler.fit_transform(X_train)
X_test = min_max_scaler.transform(X_test)
```

## Build, Compile and Fit the Model

### Building the Model

```python
from keras import backend as K
K.clear_session()
```

```python
# Input shape
X_train.shape[1]
```

```python
# build the model!
# build the model!
model = Sequential()
model.add(Dense(100, input_shape=(X_train.shape[1],), activation='relu')) # (features,)
model.add(Dropout(0.01)) 
model.add(Dense(50, activation='relu'))
model.add(Dropout(0.01))
model.add(Dense(25, activation='relu'))
model.add(Dropout(0.01))
model.add(Dense(10, activation='relu'))
model.add(Dropout(0.01)) 
model.add(Dense(1, activation='linear')) # output node
model.summary() # see what your model looks like
```

### Visualize the Shape

```python
import tensorflow as tf
tf.keras.utils.plot_model(model, show_shapes=True)
```

### Compile

```python
# compile the model
model.compile(optimizer='rmsprop', loss='mse', metrics=['mae'])
```

### Callbacks / Early Stopping

```python
# early stopping callback
es = EarlyStopping(monitor='val_loss',
                   mode='min',
                   patience=50,
                   restore_best_weights = True) 
```

### Fit the Model

```python
# fit the model!
history = model.fit(X_train, y_train,
                    validation_data = (X_test, y_test),
                    callbacks=[es], # early stopping callbacks
                    epochs=500, # set this to a big number
                    batch_size= 8, # this is a hyperparameter
                    shuffle=True, # it's a best practice to shuffle your data
                    verbose=1) # this prints the output as the model trains
```

## Evaluate The Model

### Learning Curves (Loss vs Epochs)

```python
# let's see the training and validation accuracy by epoch
history_dict = history.history
loss_values = history_dict['loss'] # you can change this
val_loss_values = history_dict['val_loss'] # you can also change this
epochs = range(1, len(loss_values) + 1) # range of X (no. of epochs)
plt.plot(epochs, loss_values, 'bo', label='Training loss')
plt.plot(epochs, val_loss_values, 'orange', label='Validation loss')
plt.title('Training and validation loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
# let's see the training and validation accuracy by epoch
history_dict = history.history
loss_values = history_dict['mae'] # you can change this
val_loss_values = history_dict['val_mae'] # you can also change this
epochs = range(1, len(loss_values) + 1) # range of X (no. of epochs)
plt.plot(epochs, loss_values, 'bo', label='Training mae')
plt.plot(epochs, val_loss_values, 'orange', label='Validation mae')
plt.title('Training and validation mae')
plt.xlabel('Epochs')
plt.ylabel('mae')
plt.legend()
plt.show()
```

### Scatter Plot Actual Vs Predicted

```python
# scatterplot of actual vs. pred
# specify the dimensions 
fig, axes = plt.subplots(1,2) # 1 row, 2 columns

# this makes the individual subplots
# Training Results
axes[0].scatter(x=y_train, y=model.predict(X_train)) #first row, first entry (left top)
axes[0].set_xlabel("Actual", fontsize=10)
axes[0].set_ylabel("Predicted",  fontsize=10)
axes[0].set_title("Training")
# add 45 deg line
x = np.linspace(*axes[0].get_xlim())
axes[0].plot(x, x, color='red')
# Validation Results
axes[1].scatter(x=y_test, y=model.predict(X_test)) # first row, second entry (right top)
axes[1].set_xlabel("Actual", fontsize=10)
axes[1].set_ylabel("Predicted",  fontsize=10)
axes[1].set_title("Validation")
# add 45 deg line
x = np.linspace(*axes[1].get_xlim())
axes[1].plot(x, x, color='red')

# tight layout
fig.tight_layout()

# show the plot
plt.show()
```

### Model Evaluation Metrics

```python
# metrics
pred = model.predict(X_test)
pred

trainpreds = model.predict(X_train)

from sklearn.metrics import mean_absolute_error

# Mean Absoulute Error
print("Mean Absolute Error for Training Dataset : ", mean_absolute_error(y_train, trainpreds)) # train
print("Mean Absolute Error for Test Dataset : ", mean_absolute_error(y_test, pred)) # test
```

```python
# R Square
print("R Square for Training Dataset : ", r2_score(y_train, trainpreds)) # train
print("R Square for Test Dataset : ", r2_score(y_test, pred)) # test
```

```python
# Mean Squared Error
print("Mean Squared Error for Training Dataset : ", mean_squared_error(y_train, trainpreds)) # train
print("Mean Squared Error for Test Dataset : ", mean_squared_error(y_test, pred)) # test
```
